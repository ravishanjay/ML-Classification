```python
import pandas as pd
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int,columns=['Gender'])
#dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset = dataset.drop("User ID",axis=1)
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset["Purchased"].value_counts()
```

```python
independent=dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
dependent=dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent,test_size=1/3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
#classifier=LogisticRegression(random_state=0)
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]} 

grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
grid.fit(X_train,Y_train)
```

```python
GridSearchCV(cv='warn', error_score='raise-deprecating',
             estimator=RandomForestClassifier(C=1.0, cache_size=200, class_weight=None, coef0=0.0,
                           decision_function_shape='ovr', degree=3,
                           gamma='auto_deprecated', kernel='rbf', max_iter=-1,
                           probability=False, random_state=None, shrinking=True,
                           tol=0.001, verbose=False),
              n_jobs=-1,
             param_grid={'C': [10, 100, 1000, 2000, 3000],
                         'gamma': ['auto', 'scale'],
                         'kernel': ['linear', 'rbf', 'poly', 'sigmoid']},
             pre_dispatch='2*n_jobs', refit=True, return_train_score=False,
             scoring='f1_weighted', verbose=3)
```

```python
y_predict = grid.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_predict)
print(cm)
```

```python
from sklearn.metrics import classification_report
Classify_Report = classification_report(Y_test,y_predict)
print(Classify_Report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,y_predict,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```
